# Preprocessing the data: Dealing with nan's in region column

```python
countries = list(set(df['Country'].values))
countries = sorted(countries)
```

```python
new_regions = []
for i in range(len(countries)):
        country_index = np.where(df['Country'] == countries[i]) # Get df of current country
        current_country_df = df.iloc[country_index]
        if all(pd.isnull(current_country_df['Region'].values)): # if all null for regions, drop country from df
            country_index = np.where(df['Country'] == countries[i])[0]
            for vals in country_index:
                try:
                    df = df.drop(index = vals)
                except KeyError:
                    pass
        else:
            region_name_pos = np.where([type(x) == str for x in current_country_df['Region'].values])[0][0]
            region = current_country_df['Region'].values[region_name_pos]
            current_country_df = current_country_df.replace(np.nan, region)
            new_regions.append(current_country_df['Region'].values)


```

```python
new_regions = list(new_regions)
k = 0
region_vals = []
for arr in new_regions:
    for vals in arr:
        region_vals.append(vals)
        k+=1
        
len(region_vals)
```

# Preprocessing the data: Final tweaks

I need to do a few things before I feed this into my decision tree:

1. Establish my classes. I will be ranking countries instead on whether their rank for a given year breaks the top 20 (1) or not (0). I will need to adjust the 'Happiness Rank' column accordingly.

2. Remove non-useful columns like 'Country', 'index', and 'Happiness Rank' (once I've dealt with it). 

3. One-hot encode 'Region', as it's non-numeric but will likely be a very important factor to include
4. Shuffle the dataset to prevent weird training

```python
df['Region'] = region_vals
```

```python
happiness_benchmark = []
for rank in df['Happiness Rank'].values:
    if rank <= 20:
        happiness_benchmark.append(1)
    else:
        happiness_benchmark.append(0)
df['Happiness Rank'] = happiness_benchmark
df=df.sample(frac=1)
```

```python
df = df.drop(columns = ['Country', 'index'])
```

```python
df = df.dropna(axis=0)
output = df['Happiness Rank']
```

# Fit to the tree

Now that I've cleaned up the data, it's ready to be fitted. I'm also going to fit with a few other classifiers to see how accurate I can get it

```python
n = len(df)
percent = int(0.8 * n)
```

```python
features = df.columns
x = df[list(features)][:percent]
y = output[:percent]
```

```python
x_test = df[list(features)][percent:]
y_test = output[percent:]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
rfc = RandomForestClassifier()
logit = LogisticRegression()
```

```python
rfc.fit(x,y)
logit.fit(x,y)
```

```python
accuracies = []
for i in np.arange(1,4): #Want to make it read-able, so beyond 4 is too much
    max_depth = i
    print('max_depth :', i)
    dtc = DecisionTreeClassifier(max_depth = max_depth)
    dtc.fit(x,y)
    pred = dtc.predict(x_test)
    print(accuracy_score(y_true = y_test, y_pred = pred))
    accuracies.append(accuracy_score(y_true = y_test, y_pred = pred))
best = np.max(accuracies)
best_depth = np.where(accuracies == best)[0][0]

dtc = DecisionTreeClassifier(max_depth = best_depth)
dtc.fit(x,y)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(40,38))
plot_tree(dtc, feature_names = features, class_names = ['Not Top 20', 'Top 20'],  # 0 -> not top 25, 1 -> top 25
          filled=True, rounded = True)
```

```python
preds_dtc = dtc.predict(x_test)
preds_rfc = rfc.predict(x_test)
preds_logit = logit.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_true = y_test, y_pred = preds_dtc)
```

```python
accuracy_score(y_true = y_test, y_pred = preds_rfc)
```

```python
accuracy_score(y_true = y_test, y_pred = preds_logit)
```

# Let's add a test sample and see how the decision tree rules it.
I've given this imaginary country values that are fairly typical for countries that perform at the top of this list -- the values here are quite similar to America's on a given year. The GDP of this country is quite high for this dataset. Critically, however, let's see what happens if the country scores poorly on freedom.

```python
data = {
       'Economy (GDP per Capita)':1.5, 'Family':1.2, 'Health (Life Expectancy)':0.8,
       'Freedom':0.2, 'Trust (Government Corruption)':0.15, 'Generosity':0.5,
       'Region_Australia and New Zealand':1, 'Region_Central and Eastern Europe':0,
       'Region_Eastern Asia':0, 'Region_Latin America and Caribbean':0,
       'Region_Middle East and Northern Africa':0, 'Region_North America':0,
       'Region_Southeastern Asia':0, 'Region_Southern Asia':0,
       'Region_Sub-Saharan Africa':0, 'Region_Western Europe':0,
}
x_test_new = pd.DataFrame(data=data, index = [0])
```

```python
dtc.predict(x_test_new) # a 0 implies that the country is *not* in the top 20 for happiness
```

As the tree would imply, even if the country is rich, if it is low scoring for freedom it is likely to still not be a top 20 country in happiness. 

A few other noteable things:

1. The GDP of the country seems to be of reasonable importance in determining its happiness
2. For countries with a GDP score below 1.3 (for reference, this value is close but slightly below what the Czech Republic scored in 2017 -- so countries below 1.3 are relatively lacking in wealth), whether the country exists in the Australia & New Zealand region is a good prediction of whether the country will be ruled as outside of the top 20 or not. 


Overall it seems to imply that:

1. Having a country in the New Zealand & Australia region is a good predictor of happiness (likely due to how well Australia and New Zealand consistently scores on Happiness rankings) even if GDP is on the low side. That said, Australia and NZ region entirely comprises the Australia and New Zealand countries, so this decision tree is more highlighting the fact that the two countries have performed well on happiness rankings, and perhaps uniquely, regardless of a lower GDP score. This tree is basically flat out saying that unless you live in Australia or New Zealand, if your GDP score is below 1.3 your country will not hit the top 20.
2. A country with a GDP score above 1.3 with a freedom score above 0.506 is more often than not classified as a top 20 country in happiness. However, high GDP scoring companies with a freedom score below 0.506 were almost eight times more likely to not be a top 20 country.
